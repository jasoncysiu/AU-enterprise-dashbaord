```python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
import numpy as np

file_path = 'Penetration_POC.xlsx'
df = pd.read_excel(file_path, sheet_name='Deal')

# Check unique owners (there are only two)
print(df['Deal owner'].value_counts())

# Group deals by Stage and Owner
stage_grouped = df.groupby(['Deal Stage', 'Deal owner']).size().reset_index(name='number')

# Desired stage order for funnel
stage_order = ['Discovery', 'Qualified', 'Piloting', 'Contracting']

# ----- Interactive Funnel Chart with Plotly -----
fig = px.funnel(
    stage_grouped,
    x='number',
    y='Deal Stage',
    color='Deal owner',
    category_orders={'Deal Stage': stage_order}
)
fig.update_layout(
    title='Deal Funnel by Stage and Deal Owner',
    xaxis_title='Number of Deals',
    yaxis_title='Stage'
)
# Save interactive chart to HTML
fig.write_html('deal_funnel_by_owner.html', include_plotlyjs='embed')

# ----- Static Funnel Chart with Matplotlib -----
pivot = stage_grouped.pivot(index='Deal Stage', columns='Deal owner', values='number').fillna(0)
pivot = pivot.reindex(stage_order)
pivot['total'] = pivot.sum(axis=1)
max_total = pivot['total'].max()

colors = {'Adrian Ku': 'cornflowerblue', 'Harry Nguyen': 'salmon'}

fig2, ax = plt.subplots(figsize=(8,6))
y_positions = np.arange(len(pivot))
for i, (stage, row) in enumerate(pivot.iterrows()):
    # Center-align each bar by calculating an offset based on the largest stage total
    x_offset = (max_total - row['total']) / 2
    x_start = x_offset
    for owner in ['Adrian Ku', 'Harry Nguyen']:
        width = row.get(owner, 0)
        if width > 0:
            ax.barh(i, width, left=x_start, color=colors[owner],
                    edgecolor='white', height=0.6, label=owner if i == 0 else '')
            ax.text(x_start + width/2, i, str(int(width)), va='center', ha='center',
                    color='white', fontsize=10, fontweight='bold')
            x_start += width

ax.set_yticks(y_positions)
ax.set_yticklabels(stage_order)
ax.set_xlabel('Number of Deals')
ax.set_title('Deal Funnel by Stage and Deal Owner')
ax.invert_yaxis()
ax.legend(title='Deal owner')
for spine in ['top', 'right']:
    ax.spines[spine].set_visible(False)

plt.tight_layout()
# plt.savefig('/home/oai/share/deal_funnel_by_owner_static.png')
```
